## Univariate Analysis

```python
#1. Amount Distribution
fig, axes = plt.subplots(nrows=6,ncols=2,figsize=(10,20))
col = ['amt','age','distance','hour','dayofweek','month']
for i in range(len(col)):
    sns.boxplot(df,x=col[i],ax=axes[i,0])
    sns.histplot(df,x=col[i],ax=axes[i,1],kde=True)
```
