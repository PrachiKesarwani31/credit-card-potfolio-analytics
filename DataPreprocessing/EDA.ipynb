## Univariate Analysis

```python
#1. Distribution of numerical variables
fig, axes = plt.subplots(nrows=6,ncols=2,figsize=(10,20))
col = ['amt','age','distance','hour','dayofweek','month']
for i in range(len(col)):
    sns.boxplot(df,x=col[i],ax=axes[i,0])
    sns.histplot(df,x=col[i],ax=axes[i,1],kde=True)
```

```python
 # Categorical variable
col = ['is_fraud','category','gender']

fig, axes = plt.subplots(nrows=3,ncols=1,figsize=(10,20))

for i in range(len(col)):
    sns.countplot(data=df,x=col[i],ax=axes[i])
```

## Bivariate Analysis

```python
#Behavioural Analysis
#1. Age by Amount
fig, axes = plt.subplots(nrows=3,ncols=1,figsize=(10,15))

s_amt = df.groupby('age_bucket',as_index=False)['amt'].sum()
s_amt = s_amt.sort_values(by='amt',ascending=False)
sns.barplot(data=s_amt,x='age_bucket',y='amt',ax=axes[0])
axes[0].set_title('Amount by Age buckets')

sns.scatterplot(data=df,x='age',y='amt',hue='is_fraud',ax=axes[1])
axes[1].set_title('Amount by Age buckets')

#2. Gender by Amount
sg = df.groupby('gender',as_index=False)['amt'].sum()
sg = sg.sort_values(by='amt',ascending=False)
sns.barplot(data=sg,x='gender',y='amt',ax=axes[2])
axes[2].set_title('Amount by Gender')
```

```python
#Merchant Analysis
fig, axes = plt.subplots(nrows=2,ncols=1,figsize=(10,10))

#1. Amount by Category
sc = df.groupby('category',as_index=False)['amt'].sum().sort_values(by='amt',ascending=False)
sns.barplot(data=sc,x='category',y='amt',ax=axes[0])
axes[0].set_title('Total Amount by Category')

#2. Fraud rate by Category
fr = df.groupby('category',as_index=False)['is_fraud'].mean().sort_values(by='is_fraud',ascending=False)
sns.barplot(data=fr,x='category',y='is_fraud',ax=axes[1])
axes[1].set_title('Fraud rate by Category')
```

```python
#Temporal Analysis
fig, axes = plt.subplots(nrows=3,ncols=2,figsize=(10,15))
col = ['hour','month','dayofweek']

for i in range(len(col)):
    s = df.groupby(col[i],as_index=False)['amt'].mean().sort_values(by='amt',ascending=False)
    f = df.groupby(col[i],as_index=False)['is_fraud'].mean().sort_values(by='is_fraud',ascending=False)
    sns.lineplot(data=s,x=col[i],y='amt',ax=axes[i,0])
    axes[i,0].set_title(f'Average Amount by {col[i]}')
    sns.lineplot(data=f,x=col[i],y='is_fraud',ax=axes[i,1])
    axes[i,1].set_title(f'Average fraud Rate by {col[i]}')
```

```python
#Fraud Analytics
#1. Fraud by Amount
fig, axes = plt.subplots(nrows=3,ncols=1,figsize=(10,15))
col = ['amt','age','distance']

for i in range(len(col)):
    sns.boxplot(data=df,x='is_fraud',y=col[i],ax=axes[i])
    axes[i].set_yscale('log')
    axes[i].set_title(f'Average {col[i]} by Fraud')
```

## Multivariate Analysis
