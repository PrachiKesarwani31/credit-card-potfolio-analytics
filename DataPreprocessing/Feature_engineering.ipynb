```python
#Category
def category_trans(category):
    s = category.strip().lower()

    if pd.isna(category):
        return np.nan
    
    else:
        try:
            for i in s:
                cat = s.split('_')
                return cat[0]
        except ValueError:
            return np.nan

df['category'] = df['category'].apply(category_trans)
```

```python
#Transaction Time Features
df['trans_date_time'] = pd.to_datetime(df['trans_date_time'])
df['hour'] = df['trans_date_time'].dt.hour
df['dayofweek'] = df['trans_date_time'].dt.dayofweek
df['month'] = df['trans_date_time'].dt.month
df.loc[df['dayofweek'].isin([5,6]), 'weekend'] = 1
df['weekend'].fillna(0,inplace=True)
```

```python
#Distance between merchant and customer
df['distance'] = np.sqrt((df['lat']-df['merch_lat'])**2+(df['lon']-df['merch_long'])**2)
```

```python
#Age Buckets
age_buckets = [20,31,51,np.inf]
age_labels = ['<30','31-50','50+']
df['age_bucket'] = pd.cut(df['age'],bins=age_buckets,labels=age_labels,right=False)
```

```python
#city population bucket
city_bins=[20,2000,3000,np.inf]
city_labels=['Low population','average population','high population']
df['city buckets'] = pd.cut(df['city_pop'],bins=city_bins,labels=city_labels,right=False)
```
